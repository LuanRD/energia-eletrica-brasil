```python
import pandas as pd
```

```python
data = pd.read_excel('../data/processed/new_clean_data.xlsx')
```

```python
data = data.drop(columns=['month', 'year'])
```

```python
data
```

```python
data.columns
```

```python
new_data = data.melt(id_vars=['date'], value_vars=['Carvao', 'Gas_Natural', 'Nuclear', 'Oleo',
       'Hidreletricas_exceto_Itaipu', 'Itaipu', 'Biomassas', 'Eolicas',
       'Solar_Fotovoltaica', 'Fora_do_SIN', 'Residuos_Industriais',
       'Renovavel', 'Nao_Renovavel', 'Hidreletricas', 'Renovaveis_sem_Hidro',
       'Total_sem_Hidro', 'Total'], var_name='Categorical_Sources', value_name='Values')
```

```python
power_bi_data = new_data.sort_values('date').reset_index(drop=True)
```

```python
power_bi_data.to_excel('../data/processed/power_bi_data.xlsx', index=False)
```

```python
power_bi_data
```

```python
power_bi_data.Categorical_Sources.unique()
```

```python
fontes = ['Carvao', 'Biomassas', 'Oleo', 'Hidreletricas_exceto_Itaipu', 'Eolicas', 'Fora_do_SIN'
           'Nuclear', 'Gas_Natural', 'Residuos_Industriais', 'Itaipu', 'Solar_Fotovoltaica']
```

```python
power_bi_data['Fontes'] = power_bi_data.apply(lambda x: x.Categorical_Sources if x.Categorical_Sources in fontes else '', axis=1)   
power_bi_data['Agrupamentos '] = power_bi_data.apply(lambda x: x.Categorical_Sources if x.Categorical_Sources not in fontes else '', axis=1)   
```
